```python
import os
```

```python
path = '/mnt/storage/target_mugshots'
target_path_glass = '/mnt/storage/multipie_viewpoint_augmentations_withglasses'
target_path_noglass = '/mnt/storage/multipie_viewpoint_augmentations_noglasses'
all_files = os.listdir(path)
all_images = [file for file in all_files if os.path.splitext(file)[1] in ['.png'] and 'mask' not in file and 'model' not in file and 'default' not in file] # jpg is either original image (duplicate) or a "model", which is a cropped face]
all_noglass = [file for file in all_images if '_____' in file]
all_glass = [file for file in all_images if 'GerretLight' in file]
```

```python
print(len(all_noglass) + len(all_glass))
print(len(all_images))
```

```python
append_this = '_il0_in0'
def rename_glass(fn):
    orig, new = fn.split('____')
    part1 = orig.replace('.jpg', '')
    orientation, ext = os.path.splitext(new.replace('GerretLight_midnightFox', ''))
    return part1 + orientation + append_this + '_1' + ext

def rename_noglass(fn):
    orig, new = fn.split('_____')
    part1 = orig.replace('.jpg', '')
    orientation, ext = os.path.splitext(new)
    return part1 + '_' + orientation + append_this + '_0' + ext
```

```python
from os.path import join

def rename_files(old, new, old_path, new_path):
    if not os.path.exists(new_path):
        os.makedirs(new_path)

    for old_name, new_name in zip(old, new):
        try:
            os.rename(join(old_path, old_name), join(new_path, new_name))
        except:
            print(old_name, "does not exist")
```

```python
all_glass_renamed = [rename_glass(elem) for elem in all_glass]
all_noglass_renamed = [rename_noglass(elem) for elem in all_noglass]
```

```python
rename_files(all_glass, all_glass_renamed, path, target_path_glass)
rename_files(all_noglass, all_noglass_renamed, path, target_path_noglass)
```
